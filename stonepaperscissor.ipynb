```python
import numpy as np

chance = int(input("Enter number of chances: "))

my_score = 0
computer_score = 0

choices = ['stone', 'paper', 'scissor']

while chance > 0:
    my_chance = input("My decision : ")
    computer_chance = np.random.choice(choices)

    print("Computer decision:", computer_chance)

    if my_chance == computer_chance:
        print("Draw")

    elif (
        (my_chance == 'stone' and computer_chance == 'scissor') or
        (my_chance == 'paper' and computer_chance == 'stone') or
        (my_chance == 'scissor' and computer_chance == 'paper')
    ):
        print("User wins")
        my_score += 1

    elif (
        (my_chance == 'stone' and computer_chance == 'paper') or
        (my_chance == 'paper' and computer_chance == 'scissor') or
        (my_chance == 'scissor' and computer_chance == 'stone')
    ):
        print("Computer wins")
        computer_score += 1

    else:
        print("Invalid choice!")
        continue

    chance -= 1

print("Final Scores")
print("User Score:", my_score)
print("Computer Score:", computer_score)

if my_score > computer_score:
    print("Overall Winner: User")
elif computer_score > my_score:
    print("Overall Winner: Computer")
else:
    print("Overall Result: Draw")



```
